```python
# Data Preprocessing Template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
print(dataset)
```

```python
# we don't apply training and test set because we don't have enough data 
# Splitting the dataset into the Training set and Test set
"""from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
"""
```

```python
#fitting linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
#visualize linear regression results
plt.scatter(X, y, color= 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Reality vs Bluff(linear regression)')
plt.xlabel('Position label')
plt.ylabel('Salary')
plt.show()
```

```python
#fitting polinomial regression 
from sklearn.preprocessing import PolynomialFeatures
poly_reg =  PolynomialFeatures(degree = 10)
X_poli = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poli, y)
```

```python
#visualize polinomial regression results
plt.scatter(X, y, color= 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Reality vs Bluff(polynomial regression)')
plt.xlabel('Position label')
plt.ylabel('Salary')
plt.show()
```

```python
#Predicting a new result with Linear Regression
lin_reg.predict([[6.5]])
```

```python
#Predicting a new result with Polinomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

```python
# Ridge Regression (L2 Regularization) 
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
# Create a Ridge regression model with a specified alpha (lambda)
ridge_reg = Ridge(alpha=1000000)
# vary alpha to get various result
```

```python
# Fit the model to your training data
# ridge_reg.fit(X_train, y_train)
ridge_reg.fit(X, y)
```

```python
# Make predictions on the test data
# y_pred = ridge_reg.predict(X_test)
y_pred = ridge_reg.predict(X)
```

```python
# Evaluate the model's performance (e.g., mean squared error)
# mse = mean_squared_error(y_test, y_pred)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values

# Fitting linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Fitting polynomial regression
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

# Fitting ridge regression with regularization
ridge_reg = Ridge(alpha=1.0)  # You can adjust the alpha value as needed
ridge_reg.fit(X_poly, y)

# Visualize linear regression results
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Reality vs Bluff (Linear Regression)')
plt.xlabel('Position label')
plt.ylabel('Salary')
plt.show()

# Visualize polynomial regression results
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Reality vs Bluff (Polynomial Regression)')
plt.xlabel('Position label')
plt.ylabel('Salary')
plt.show()

# Visualize ridge regression results
plt.scatter(X, y, color='red')
plt.plot(X, ridge_reg.predict(X_poly), color='blue')
plt.title('Reality vs Bluff (Ridge Regression)')
plt.xlabel('Position label')
plt.ylabel('Salary')
plt.show()


```

```python
# Predicting a new result with Linear Regression
lin_reg.predict(np.array([[6.5]]))
```

```python
# Predicting a new result with Polynomial Regression
lin_reg_2.predict(poly_reg.fit_transform(np.array([[6.5]])))
```

```python
# Predicting a new result with Ridge Regression
ridge_reg.predict(poly_reg.fit_transform(np.array([[6.5]])))
```
