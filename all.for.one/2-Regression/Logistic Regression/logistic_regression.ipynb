```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate some hypothetical data
np.random.seed(0)
hours_studied = np.sort(np.random.uniform(0, 10, 100))  # X-axis: Hours studied
pass_exam = np.where(hours_studied < 5, 0, 1)  # Y-axis: 0 for fail, 1 for pass
```

```python
# Define the sigmoid function
def sigmoid(x, a, b):
    return 1 / (1 + np.exp(-a * (x - b)))
```

```python
# Fit the sigmoid function to the data
params, covariance = curve_fit(sigmoid, hours_studied, pass_exam)

# Extract the parameters
a, b = params

# Generate points for the sigmoid curve
x_curve = np.linspace(0, 10, 100)
y_curve = sigmoid(x_curve, a, b)
```

```python
# Plot the data and the fitted sigmoid curve
plt.figure(figsize=(8, 6))
plt.scatter(hours_studied, pass_exam, label='Data', color='blue')
plt.plot(x_curve, y_curve, label='Fitted Sigmoid Curve', color='red')
plt.xlabel('Hours Studied')
plt.ylabel('Pass Exam (1 = Yes, 0 = No)')
plt.legend()
plt.title('Sigmoid Function Fit to Data')
plt.grid(True)
plt.show()
```

```python
# Predict using the sigmoid function
new_hours = 4.98  # New data point
predicted_pass = sigmoid(new_hours, a, b)
print(f"Predicted Probability of Passing: {predicted_pass:.2f}")
```
